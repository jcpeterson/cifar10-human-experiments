```python
from jutils import *
import numpy as np
```

```python
counts_by_image = load_cifar10h_labels()

for orig_counts in counts_by_image:

    
    for n_samples in range(2, 24):
    
        counts = orig_counts.copy()
        orig_probs = counts / counts.sum()

        probs = orig_probs.copy()

        samples = []
        while counts.sum() > 0.1:
            pick = np.random.multinomial(1, probs)
            counts -= pick.copy()
            probs = counts / counts.sum()

            samples.append(pick.copy())
            if len(samples) == 25:
                break
        new_probs = np.mean(samples, axis=0)
        print(np.corrcoef(orig_probs, new_probs))
        break
```
