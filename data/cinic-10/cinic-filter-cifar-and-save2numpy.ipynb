```python
import os, sys
import numpy as np
from scipy import misc

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# maintain this order for 0-9 labels
categories = [
    'airplane',
    'automobile',
    'bird',
    'cat',
    'deer',
    'dog',
    'frog',
    'horse',
    'ship',
    'truck'
]
# cinic splits are in these folders
data_folders = ['train','test','valid']
```

```python
# check counts for each split / category (find cifar images too)
for data_folder in data_folders:
    folder_total = 0
    for cat in categories:
        path = os.path.join(data_folder, cat)
        files = os.listdir(path)
        total_len = len(files)
        cifar_len = len([x for x in files if 'cifar' in x])
        folder_total += total_len - cifar_len
        print(data_folder, cat, total_len, cifar_len)
    print(folder_total)
    print('')
```

```python
# read in images and store in numpy arrays
images = []
for data_folder in data_folders:
    print(data_folder)
    for cat in categories:
        path = os.path.join(data_folder, cat)
        fns = [x for x in os.listdir(path) if '.png' in x]
        # remove cifar images
        fns = [x for x in os.listdir(path) if 'cifar' not in x]
        print('--', cat, len(fns))
        sys.stdout.flush()
        for i, fn in enumerate(fns):
            image = misc.imread(os.path.join(path, fn))
            images.append(image)
            if (i % 1000 == 0) and i != 0:
                print(i)
```

```python
# fix black/white images
images_fixed = []
count = 0
for image in images:
    if image.shape != (32,32,3):
        bw = image.copy()
        color_im = np.zeros((32,32,3), dtype='uint8')
        for chan in range(3):
            color_im[:,:,chan] = bw.copy()
        images_fixed.append(color_im.copy())
        count += 1
    else:
        images_fixed.append(image.copy())
print(count)
```

```python
np.save('cinic-no-cifar-images-tmp.npy', np.array(images_fixed))
```

```python
# create labels for 
labels = []
for data_folder in data_folders:
    print(data_folder)
    for c, cat in enumerate(categories):
        path = os.path.join(data_folder, cat)
        fns = [x for x in os.listdir(path) if '.png' in x]
        # remove cifar images
        fns = [x for x in os.listdir(path) if 'cifar' not in x]
        print('--', cat, len(fns))
        sys.stdout.flush()
        for i, fn in enumerate(fns):
            labels.append(c)
```
